```python
from sklearn.datasets import load_diabetes
import pandas as pd
import seaborn as sns
sns.set()
diabetes_data = load_diabetes()
print(diabetes_data["DESCR"])
df=pd.DataFrame(diabetes_data.data, columns=diabetes_data.feature_names)
df
```

```python
from sklearn import preprocessing

x = df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled, columns=df.columns)

df
```

```python
# Below is Linear Regression for TCells and LDLs. TCells used as the dependant variable and LDLs the independant variable to find how high LDLs may impact TCells
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df.loc[:,['s2','s1']]
```

```python
plt.scatter(df.s2, df.s1);
```

```python
modelBloodSugarTriglicerides= linear_model.LinearRegression()
modelBloodSugarTriglicerides.fit(df[['s2']], df.s1)
```

```python
modelBloodSugarTriglicerides.coef_
```

```python
modelBloodSugarTriglicerides.intercept_
```

```python
modelBloodSugarTriglicerides.score(df[['s2']], df.s1)
```

```python
y_fit = modelBloodSugarTriglicerides.predict (df[['s2']])
plt.scatter(df.s2, df.s1)
plt.plot(df[['s2']], y_fit);
```
